```python
import pandas as pd

# Read the Excel file, skipping the first row
df = pd.read_excel('SL Travel Itinerary - Tagged.xlsx', skiprows=1)

# Set the column names
df.columns = ['ID'] + df.columns[1:].tolist()  # Change the first column to 'ID'

# Verify the column names
print(df.columns)
# populate the ID column with UUID
import uuid
df['ID'] = [str(uuid.uuid4()) for _ in range(len(df))]
df
```

```python
df.fillna('NA', inplace=True)
# get the first row of DF
first_row = df.iloc[0]
# turn the row into a dictionary
first_row_dict = first_row.to_dict()
first_row_dict

# create a list of dictionaries
data = []
for i in range(len(df)):
    data.append(df.iloc[i].to_dict())
```

```python
import requests
import json

url = "https://kq17idrqs9.execute-api.us-east-1.amazonaws.com/prod/upload"
items = [first_row_dict]
response = requests.post(url, json=items)
print(f"Status Code: {response.status_code}, Response: {response.json()}")

# Convert list of dictionaries to JSON
```

```python
response = requests.post(url, json=data)
print(f"Status Code: {response.status_code}, Response: {response.json()}")
```
